```python
import pandas as pd

tochi_train = pd.read_csv('../data/train_genba.tsv', sep='\t')
build_train = pd.read_csv('../data/train_goto.tsv', sep='\t')
train = pd.merge(tochi_train, build_train, on="pj_no")

tochi_test = pd.read_csv('../data/test_genba.tsv', sep='\t')
build_test = pd.read_csv('../data/test_goto.tsv', sep='\t')
test = pd.merge(tochi_test, build_test, on="pj_no")
```

```python
unique = list(set(list(test['jukyo'])) - set(list(train['jukyo'])))
unique
```

```python
import os
import requests
from bs4 import BeautifulSoup
import time
from tqdm import tqdm

#os.remove('longlat.txt')

url = 'http://www.geocoding.jp/api/'
latlons = []
for i, address in enumerate(unique):
    payload = {"v": 1.1, 'q': address}
    r = requests.get(url, params=payload)
    ret = BeautifulSoup(r.content,'lxml')
    try:
        lat = ret.find('lat').string
        lon = ret.find('lng').string
        latlon = lat+','+lon
    except Exception:
        latlon = 'NaN'
    print('Iteration {} of {}. Geocode of {} is {}.'.format(i, len(unique), address, latlon))
    with open("longlat.txt",'a+') as f:
        f.write(address+','+latlon+'\n')
    time.sleep(1)
        
#train['location'] = pd.read_csv('longlat.txt')
#train['point'] = train['location'].apply(lambda loc: tuple(loc.point) if loc else None)
```
